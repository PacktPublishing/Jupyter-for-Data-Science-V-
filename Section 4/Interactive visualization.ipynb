```python
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_notebook 
import numpy as np
import pandas as pd
import scipy.special 
from_counts = np.load("./files/from_counts.npy")
df = pd.DataFrame({'Votes':from_counts})
print(df.head())
df.plot.hist(bins=12)
p1 = figure(title="How Many Votes Made by Users",tools="save", background_fill_color="white")
hist, edges = np.histogram(df['Votes'].values, bins=12)
p1.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="pink", line_color="pink") 
p1.legend.location = "center_right" 
p1.legend.background_fill_color = "darkgrey" 
p1.xaxis.axis_label = 'a'
p1.yaxis.axis_label = 'Count(A)' 
output_notebook()
show(gridplot(p1, ncols=1, plot_width=400, plot_height=400, toolbar_location=None))
```
